# Custom Setting Pandas
- ubah jumlah kolom yang diperlihatkan `pd.set_option("display.max.columns", None)`
- ubah jumlah desimal poin yang diperlihatkan dengan `pd.set_option("display.precision", 2)`

```python
pd.set_option("display.max.columns", None)
pd.set_option("display.precision", 2)
```

# Persiapan data
- cek data kosong (null) lalu di hapus atau diisi yang kosong ini
    - `dropna()` untuk hapus data kosong
    - `fillna()` untuk isi data kosong
- cek invalid data, data aneh seperti game nba skor nya 0

```python
nba_dropped = nba.dropna()
nba_dropped.info()
```

```python
nba_dropped = nba.dropna(axis=1)
nba_dropped.info()
```
